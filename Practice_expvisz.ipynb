Data exploration - pandas 


```python
import pandas as pd 

#load clean file
file = "D:\Data_Science\Data_Exploration_Visulaization\Practice data\cleaned_data.csv"
df = pd.read_csv(file)

#verify 
df.head(31)
```

```python
df.info()
```

```python
#reorder columns to make life easier
df = df[["Date","Year","Month","Day","Duration","Pulse","Maxpulse","Calories"]]
```

```python
#verify 
df.info()
```

```python
#summary of data 
df.describe()
```

```python

duration_count = df.groupby("Duration").size()
print(duration_count)

```

```python
#How many 30,45,60 min workout sessions for dec? - plot bar chart? 
duration_count = df.groupby("Duration").size()

duration_count.index = duration_count.index.astype(str) + " minutes"

# Plot pie chart using Pandas only
duration_count.plot.pie(
    autopct='%1.1f%%',
    figsize=(3, 3),
    startangle=90,
    title='Workout sessions in Dec')

```

```python
pulse_data = df.iloc[:,2:4]
print(pulse_data)

```

```python
pulse_data.plot(kind = "box",title = "Pulse data")
```

```python
#avg calories burnt for each duration 
# x = duration - 30 45 -60 mins
# y = avg calories burnt - group by and avg 

calories_per_session = df.groupby("Duration")["Calories"].mean()

calories_per_session.plot(x="Duration",y="Calories",kind="bar",title="Avg calories burnt")
```

```python
#avg pulse in each workout session 
#x-duration
#y - avg pulse 
avgpulse_per_session = df.groupby("Duration")["Pulse"].mean()
avgpulse_per_session.plot(x="Duration",y="Pulse",kind="bar",title="Avg pulse rate for each workout session") 
```

```python
df.plot(x="Day",y="Calories",kind ="line")
```

```python
#avg of cal and pulse for each session
df_calpulse_persession = df.groupby("Duration")[["Calories","Pulse"]].mean()
df_calpulse_persession.plot(kind ="bar",title="avg cal and pulse for each workout session")
```
